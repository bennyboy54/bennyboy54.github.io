```python
import pandas as pd
import numpy as np
food_atlas = pd.ExcelFile('FoodEnvironmentAtlas Mergable.xlsx')
list_dataframes = []

for sheet in food_atlas.sheet_names:
  list_dataframes.append(food_atlas.parse(sheet))

combined_atlas = pd.concat(list_dataframes, ignore_index= True)
final_df = combined_atlas.groupby('County').agg(lambda x: x.iloc[0] if x.dtype == 'O' else x.sum()).reset_index()
#This cell combines all of the Excel sheets into one dataframe
```

```python
final_df.loc[final_df["GROC11"]==0]#All of the counties that did not have any grocery stores in 2011
```

```python
print(final_df[["GROC11", "GROCPTH11"]].describe())#stats of number of grocery stores in counties and grocery stores per 1000 people per county

print(final_df.loc[final_df["GROC11"].idxmin()])#This county has 0 grocery stores
```
